```sql
SELECT *, CONCAT(driver_ref,'-',code) AS new_driver_ref
FROM drivers
```

```sql
SELECT *, SPLIT(name, ' ')[0] AS forename, SPLIT(name, ' ')[1] AS surname
FROM drivers
```

```sql
SELECT *, date_format(dob,'dd-MM-yyyy') FROM drivers
```

```sql
SELECT MAX(dob) FROM drivers
```

```sql
SELECT * FROM drivers WHERE dob = '2000-05-11'
```

```sql
SELECT COUNT(*) FROM drivers WHERE nationality = 'British'
```

```sql
SELECT COUNT(*), nationality FROM drivers GROUP BY nationality
```

```sql
SELECT COUNT(*), nationality FROM drivers GROUP BY nationality HAVING COUNT(*)>100 ORDER BY nationality
```

```sql
 SELECT nationality,name,dob, RANK() OVER(PARTITION BY nationality ORDER BY dob) AS age_rank
 FROM drivers ORDER BY nationality,age_rank
```
