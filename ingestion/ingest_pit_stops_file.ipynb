##ingest pit stops json

### 1. read data json

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType, DateType
```

```python
pit_stop_schemas = StructType(fields=[
    StructField("raceId", IntegerType(), False),
    StructField("driverId", IntegerType(), True),
    StructField("stop", StringType(), True),
    StructField("lap", IntegerType(), True),
    StructField("time", StringType(), True),
    StructField("duration", StringType(), True),
    StructField("milliseconds", IntegerType(), True)
])
```

```python
pit_stops_df=spark.read.schema(pit_stop_schemas).option("multiline",True).json("abfss://raw@formula1dlbtg.dfs.core.windows.net/pit_stops.json")
```

```python
display(pit_stops_df)
```

###2.rename

```python
from pyspark.sql.functions import current_timestamp
```

```python
final_df=pit_stops_df.withColumnRenamed("raceId","race_id") \
.withColumnRenamed("driverId","driver_id") \
    .withColumn("ingestion_date",current_timestamp())
display(final_df)
```

```python
final_df.write.mode("overwrite").parquet("abfss://processed@formula1dlbtg.dfs.core.windows.net/pit_stops")
```
