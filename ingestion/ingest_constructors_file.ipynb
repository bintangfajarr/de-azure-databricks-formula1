### ingest constructors file json

####1. read file json

```python
constructors_schema="constructorId INT, constructorRef STRING, name STRING, nationality STRING, url STRING"
```

```python
constructor_df = spark.read.schema(constructors_schema).json("abfss://raw@formula1dlbtg.dfs.core.windows.net/constructors.json")
```

```python
constructor_df.printSchema()
```

```python
display(constructor_df)
```

drop col

```python
from pyspark.sql.functions import col
```

```python
constructor_dropped_df=constructor_df.drop(col("url"))
```

rename columns and add ingestion


```python
from pyspark.sql.functions import current_timestamp
```

```python
constructor_final_df=constructor_dropped_df.withColumnRenamed("constructorId","constructor_id") \
.withColumnRenamed("constructorRef","constructor_ref") \
.withColumn("ingestion_date",current_timestamp())
```

write output

```python
constructor_final_df.write.mode("overwrite").parquet("abfss://processed@formula1dlbtg.dfs.core.windows.net/constructors")
```

```python
df=spark.read.parquet("abfss://processed@formula1dlbtg.dfs.core.windows.net/constructors")
display(df)
```
